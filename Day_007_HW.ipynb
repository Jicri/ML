```python
import pandas as pd
import numpy as np
import os

data_path = "data/"
train_path = os.path.join(data_path,'titanic.train.csv')
test_path = os.path.join(data_path,'titanic.test.csv')

app_train = pd.read_csv(train_path)
app_test = pd.read_csv(test_path)
app_train
```

```python
app_test
```

```python
train_Y = app_train['Survived']
ids = app_test['PassengerId']
df_train = app_train.drop(['PassengerId','Survived'],axis=1)
df_test = app_test.drop(['PassengerId'],axis=1)
df = pd.concat([df_train,df_test])
df
```

```python
dtype_df = df.dtypes.reset_index()
dtype_df.columns = ['count','columns Type']
dtype_df
```

```python
dtype_df = dtype_df.groupby('columns Type').aggregate('count').reset_index()
dtype_df
```

```python
int_features = []
float_features = []
object_features = []

for dtype,feature in zip(df.dtypes,df.columns):
    if dtype == 'float64':
        float_features.append(feature)
    elif dtype == 'int64':
        int_features.append(feature)
    else :
        object_features.append(feature)

print("int feature :",int_features,"\n")
print("float feature :",float_features,"\n")
print("object feature :",object_features)
df[int_features]
```

```python
df[object_features].nunique()
```

```python
df[object_features].mean()
```

```python
type_data = {"int feature":int_features,"float feature":float_features,"object feature":object_features}

for d in type_data:
    print(d,'得平均值',"\n","\n",df[type_data[d]].mean())
    print(d,'得最大值',"\n","\n",df[type_data[d]].max())
    print(d,'得特徵值',"\n","\n",df[type_data[d]].nunique())    
```

```python
def type_process(data,group):
    print('Mean()')
    print(data[group].mean())
print(type_process(df,int_features))
```
